```python
import sys
from math import*
from collections import deque
row, col = 1000, 1000
import turtle
#screen = turtle.Screen()
#screen.title()
def check(a, b):
    change = [-1]*30
    for i in range(0, len(a)):
        k = ord(a[i]) - ord('a')
        if change[k] == -1:
            change[k] = ord(b[i]) - ord('a')
        else:
            if change[k] != ord(b[i]) - ord('a'): return 0
    ok = [0]*30
    for i in range(0, 27):
        if change[i] == -1: continue
        if ok[change[i]] == 1: return 0
        ok[change[i]] = 1
    return 1
if __name__ == '__main__':
    try:
        sys.stdin = open("task.in", "r")
        sys.stdout = open("task.out", "w")
    except FileNotFoundError:
        pass
    a, b = map(str, input().split())
    if check(a, b) == 1: print("true")
    else: print("false")
```
