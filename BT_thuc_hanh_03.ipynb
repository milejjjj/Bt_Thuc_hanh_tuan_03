```python
#bai11
import sys
from math import*
from collections import deque
row, col = 1000, 1000
import turtle
#screen = turtle.Screen()
#screen.title()
def deu(x, y, z):
    if(x == y and y == z): return 1
    return 0
def can(x, y, z):
    if(x == y or x == z or y == z): return 1
    return 0
if __name__ == '__main__':
    try:
        sys.stdin = open("task.in", "r")
        sys.stdout = open("task.out", "w")
    except FileNotFoundError:
        pass
    a, b, c = map(int, input().split())
    if(a + b > c and a + c > b and b + c > a):
        if deu(a, b, c): print("Tam giác đều")
        elif can(a, b, c): print("Tam giác cân")
        else: print("Tam giác thường")
    else: print("Không phải tam giác")
```

```python
#bai16
import sys
from math import*
from collections import deque
row, col = 1000, 1000
import turtle
#screen = turtle.Screen()
#screen.title()

if __name__ == '__main__':
    try:
        sys.stdin = open("task.in", "r")
        sys.stdout = open("task.out", "w")
    except FileNotFoundError:
        pass
    n = float(input())
    val = int(n)
    tmp = float(val)
    if tmp == n: print(val, val, val, sep=' ')
    else:
        dis = abs(n - tmp)
        if n < 0:
            if dis < 0.5:  print(val, val - 1, val)
            else: print(val, val - 1, val - 1)
        else:
            if(dis < 0.5): print(val + 1, val, val)
            else: print(val + 1, val, val + 1)
```
