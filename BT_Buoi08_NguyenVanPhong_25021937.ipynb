```python
class Cylinder:
    def __init__(self, r, h):
        self.r = r
        self.h = h
    def calcS(self):
        ans = 2*3.14*self.r*(self.h + self.r)
        print("%.2f"%ans)
    def calcV(self):
        ans = 3.14*self.r*self.r*self.h
        print("%.2f"%ans)
r, h = map(float, input().split())
tmp = Cylinder(r, h)
tmp.calcS()
tmp.calcV()
```

```python
class Calculator:
    def __init__(self, a, b):
        self.a, self.b = a, b
    def add(self):
        return self.a + self.b
    def subtract(self):
        return self.a - self.b
    def product(self):
        return self.a*self.b
    def div(self):
        return self.a/self.b
    def pow(self):
        return self.a**self.b
    def mod(self):
        return self.a%self.b
def main():
    print("_____Chương trình máy tính bỏ túi_____")
    try:
        a = float(input("Nhập số thứ nhất: "))
        b = float(input("Nhập số thứ hai: "))
    except ValueError:
        print("Vui lòng chỉ nhập số!")
        return
    calc = Calculator(a, b)
    while True:
        print("\n--- MENU ---")
        print("1. Cộng (add)")
        print("2. Trừ (subtract)")
        print("3. Nhân (product)")
        print("4. Chia (divide)")
        print("5. Lũy thừa (power)")
        print("6. Chia lấy dư (mod)")
        print("7. Nhập lại 2 số mới (set_numbers)")

        choice = input("Chọn phép tính (1-7): ")
        if choice == '1':
            print(f"Kết quả cộng: {calc.add()}")
        elif choice == '2':
            print(f"Kết quả trừ: {calc.subtract()}")
        elif choice == '3':
            print(f"Kết quả nhân: {calc.product()}")
        elif choice == '4':
            print(f"Kết quả chia: {calc.div()}")
        elif choice == '5':
            print(f"Kết quả lũy thừa: {calc.pow()}")
        elif choice == '6':
            print(f"Kết quả chia dư: {calc.mod()}")
        elif choice == '7':
            try:
                new_a = float(input("Nhập số thứ nhất mới: "))
                new_b = float(input("Nhập số thứ hai mới: "))
                tmp = Calculator(new_a, new_b)
                calc = tmp
            except ValueError:
                print("Lỗi: Vui lòng nhập số hợp lệ.")
        else:
            print("Lựa chọn không hợp lệ!")
        exit_check = input("\nBạn có muốn thoát? (Gõ 'yes' để thoát, phím khác để tiếp tục): ")
        if exit_check in ["yes", "YES", "Yes"]:
            print("Đã thoát chương trình. Tạm biệt!")
            break

if __name__ == "__main__":
    main()
```

```python
class ShoppingCart:
    def __init__(self):
        self.cart = []

    def them_san_pham(self, ten, gia):
        san_pham = {"name": ten, "price": gia}
        self.cart.append(san_pham)
        print(f"-> Đã thêm '{ten}' với giá {gia} vào giỏ hàng.")

    def xoa_san_pham(self, ten):
        found = False
        for sp in self.cart:
            if sp["name"] == ten:
                self.cart.remove(sp)
                print(f"-> Đã xóa sản phẩm '{ten}' khỏi giỏ hàng.")
                found = True
                break
        if not found:
            print(f"-> Không tìm thấy sản phẩm '{ten}' trong giỏ hàng.")

    def kiem_tra_rong(self):
        if len(self.cart) == 0:
            print("-> Giỏ hàng đang TRỐNG.")
            return True
        else:
            print("-> Giỏ hàng KHÔNG trống.")
            return False

    def tinh_tong_tien(self):
        tong = 0
        for sp in self.cart:
            tong += sp["price"]
        print(f"-> Tổng tiền giỏ hàng là: {tong}")
        return tong

    def hien_thi_gio_hang(self):
        if len(self.cart) == 0:
            print("-> Giỏ hàng trống, chưa có sản phẩm nào.")
        else:
            print("\n--- DANH SÁCH SẢN PHẨM TRONG GIỎ ---")
            for i, sp in enumerate(self.cart, 1):
                print(f"{i}. Tên: {sp['name']}, Giá: {sp['price']}")
            print("------------------------------------")

    def xoa_toan_bo(self):
        self.cart = []
        print("-> Đã xóa toàn bộ giỏ hàng.")

def main():
    gio_hang = ShoppingCart()

    while True:
        print("\n================ MENU ================")
        print("2. Thêm sản phẩm vào giỏ")
        print("3. Xóa sản phẩm khỏi giỏ (theo tên)")
        print("4. Kiểm tra giỏ hàng rỗng")
        print("5. Tính tổng tiền")
        print("6. Hiển thị giỏ hàng")
        print("7. Xóa toàn bộ giỏ hàng")
        print("======================================")

        try:
            lua_chon = int(input("Nhập lựa chọn của bạn (2-7): "))
        except ValueError:
            print("Vui lòng nhập một con số hợp lệ.")
            continue

        if lua_chon == 2:
            ten = input("Nhập tên sản phẩm: ")
            try:
                gia = float(input("Nhập giá sản phẩm: "))
                gio_hang.them_san_pham(ten, gia)
            except ValueError:
                print("Lỗi: Giá phải là một con số.")

        elif lua_chon == 3:
            ten = input("Nhập tên sản phẩm cần xóa: ")
            gio_hang.xoa_san_pham(ten)

        elif lua_chon == 4:
            gio_hang.kiem_tra_rong()

        elif lua_chon == 5:
            gio_hang.tinh_tong_tien()

        elif lua_chon == 6:
            gio_hang.hien_thi_gio_hang()

        elif lua_chon == 7:
            gio_hang.xoa_toan_bo()

        else:
            print("Lựa chọn không hợp lệ. Vui lòng chọn từ 2 đến 7.")
        thoat = input("\nCó muốn thoát khỏi chương trình không? (nhập 'yes' để thoát): ")
        if thoat.lower() == "yes":
            print("Cảm ơn bạn đã sử dụng chương trình. Tạm biệt!")
            break

if __name__ == "__main__":
    main()
```

```python
class Rectangle:
    def __init__(self, w, h):
        self.w = w
        self.h = h
    def area(self):
        ans = self.w*self.h
        print("%.2f"%ans, end = ' ')
    def perimeter(self):
        ans = 2*(self.w + self.h)
        print("%.2f"%ans)
    def scale(self, k):
        self.h = self.h*k
        self.w = self.w*k
w, h, k = map(float, input().split())
tmp = Rectangle(w, h)
tmp.scale(k)
tmp.area()
tmp.perimeter()
```

```python
from math import *
def gcd(a, b):
    if b == 0:
        return a
    return gcd(b, a%b)
def lcm(a, b):
    tmp = a*b
    return tmp//gcd(a, b)
class Fraction:
    def __init__(self, num, den):
        tmp = gcd(num, den)
        self.num = num//tmp
        self.den = den//tmp
    def show(self):
        print(self.num, '/', self.den, sep = '')
    def add(self, other):
        ans = Fraction(1, 1)
        ans.den = lcm(self.den, other.den)
        ans.num = self.num*(ans.den//self.den) + other.num*(ans.den//other.den)
        ans = Fraction(ans.num, ans.den)
        return ans
    def sub(self, other):
        other = Fraction(-other.num, other.den)
        ans = self.add(other)
        return ans
    def mul(self, other):
        ans = Fraction(self.num*other.num, self.den*other.den)
        return ans
    def div(self, other):
        ans = Fraction(self.num*other.den, self.den*other.num)
        return ans
a, b, op, c, d = map(str, input().split())
a = int(a)
b = int(b)
c = int(c)
d = int(d)
x = Fraction(a, b)
y = Fraction(c, d)
if op == "+": x.add(y).show()
if op == "-": x.sub(y).show()
if op == "*": x.mul(y).show()
if op == "/": x.div(y).show()
```

```python
class BankAccount:
    def __init__(self,name, cost):
        self.cost = cost
    def deposit(self, val):
        self.cost += val
    def withdraw(self, val):
        if self.cost >= val:
            self.cost -= val
    def show(self):
        print("%.2f"% self.cost)
name, cost = map(str, input().split())
cost = float(cost)
tmp = BankAccount(name, cost)
n = int(input())
for i in range(n):
    type, val = map(str, input().split())
    val = float(val)
    if type == "DEPOSIT": tmp.deposit(val)
    else: tmp.withdraw(val)
tmp.show()
```

```python
class student:
    score = []
    def __init__(self):
        self.score = []
    def add(self, sub, val):
        self.score.append(val)
    def avg(self):
        ans = 0
        for it in self.score:
            ans += it
        return ans/len(self.score)
    def rank(self):
        tmp = self.avg()
        if tmp >= 8: print("Excellent")
        elif tmp >= 6.5: print("Good")
        elif tmp >= 5: print("Average")
        else:  print("Poor")
name = input()
n = int(input())
tmp = student()
for i in range(n):
    x, y = map(str, input().split())
    y = float(y)
    tmp.add(x, y)
print(name, "%.2f"%tmp.avg(), end = ' ')
tmp.rank()

```

```python
import sys
class Library:
    def __init__(self):
        self.books = []
        self.author_count = {}
        self.year_count = {}
    def add(self, title, author, year):
        if author not in self.author_count:
            self.author_count[author] = 0
        if year not in self.year_count:
            self.year_count[year] = 0
        self.author_count[author] += 1
        self.year_count[year] += 1
def solve():
    n = int(input())
    lib = Library()
    ans = []
    for i in range(0, n):
        line = input()
        if line.startswith("ADD"):
            content = line[4:].split(';')
            title = content[0]
            author = content[1]
            year = content[2]
            lib.add(title, author, year)
        elif line.startswith("COUNTYEAR"):
            year = line.split()[1]
            if year not in lib.year_count:
                lib.year_count[year] = 0
            ans.append(lib.year_count[year])
        elif line.startswith("COUNT"):
            author = line[6:].strip()
            if author not in lib.author_count:
                lib.author_count[author] = 0
            ans.append(lib.author_count[author])
    print(ans)
if __name__ == "__main__":
    solve()

```

```python
from math import*
class point:
    def __init__(self, a, b, c):
        self.x, self.y, self.z = a, b, c
    def vctpt(self, other1, other2):
        ab = point(other1.x - self.x, other1.y - self.y, other1.z - self.z)
        ac = point(other2.x - self.x, other2.y - self.y, other2.z - self.z)
        ans = point(0, 0, 0)
        ans.x = ab.y*ac.z - ac.y*ab.z
        ans.y = ab.z*ac.x - ab.x*ac.z
        ans.z = ab.x*ac.y - ac.x*ab.y
        return ans
    def calcgoc(self, B, C, D):
        n1 = self.vctpt(B, C)
        n2 = B.vctpt(C, D)
        cos = abs(n1.x*n2.x + n1.y*n2.y + n1.z*n2.z)/(sqrt(n1.x*n1.x + n1.y*n1.y + n1.z*n1.z)*sqrt(n2.x*n2.x + n2.y*n2.y + n2.z*n2.z))
        ans = degrees(acos(cos))
        print("%.2f"%ans)
x1, y1, z1 = map(float, input().split())
x2, y2, z2 = map(float, input().split())
x3, y3, z3 = map(float, input().split())
x4, y4, z4 = map(float, input().split())
A = point(x1, y1, z1)
B = point(x2, y2, z2)
C = point(x3, y3, z3)
D = point(x4, y4, z4)
A.calcgoc(B, C, D)
```

```python
class ComplexNumber:
    def __init__(self, re, im):
        self.re, self.im = re, im
    def multiply(self, others):
        ans = ComplexNumber(0, 0)
        a, b, c, d = self.re, self.im, others.re, others.im
        ans.re = a*c - b*d
        ans.im = a*d + b*c
        print(ans.re, ans.im)
a, b = map(float, input().split())
c, d = map(float, input().split())
tmp1 = ComplexNumber(a, b)
tmp2 = ComplexNumber(c, d)
tmp1.multiply(tmp2)
```

```python
from math import *
class Triangle:
    def __init__(self, a, b, c):
        self.a, self.b, self.c = a, b, c
    def check(self):
        a, b, c = self.a, self.b, self.c
        if a + b <= c or a + c <= b or b + c <= a: return -1
        return 1
    def getArea(self):
        a, b, c = self.a, self.b, self.c
        p = (a + b + c)/2
        ans = sqrt(p*(p - a)*(p - b)*(p - c))
        print("%.2f"%ans)
a, b, c = map(int, input().split())
tmp = Triangle(a, b, c)
if tmp.check() == -1: print("invalid")
else: tmp.getArea()

```
