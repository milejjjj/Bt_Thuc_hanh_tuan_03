```python
import sys
from math import*
from collections import deque
row, col = 1000, 1000
import turtle
#screen = turtle.Screen()
#screen.title()
def change(tmp):
    last = 0
    cnt = 0
    pos = []
    dic = {}
    lis = []
    for i in range(0, len(tmp)):
        if tmp[i] == "'":
            cnt += 1
            if cnt %2 == 0:
                string = tmp[(last + 1):i]
                lis.append(string)
                dic[string] = 0
            last = i
        if tmp[i] == ",": pos.append(i)
    cnt = 0
    for i in range(0, len(tmp)):
        if tmp[i] == ":":
            cnt += 1
            if len(pos) >= 1:
                dic[lis[cnt - 1]] += int(tmp[(i + 2):pos[0]])
                pos.remove(pos[0])
            else:
                dic[lis[cnt - 1]] += int(tmp[(i + 2):(len(tmp) - 1)])
    return dic

if __name__ == '__main__':
    try:
        sys.stdin = open("task.in", "r")
        sys.stdout = open("task.out", "w")
    except FileNotFoundError:
        pass
    s = input()
    t = input()
    dics = change(s)
    dict = change(t)
    dicans = {}
    for key, val in dics.items():
        if key not in dicans: dicans[key] = 0
        dicans[key] += val
    for key, val in dict.items():
        if key not in dicans: dicans[key] = 0
        dicans[key] += val
    lis = []
    for key in dicans.keys():
        lis.append(key)
    lis.sort()
    dicres = {}
    for it in lis:
        dicres[it] = 0
        dicres[it] = dicans[it]
    print(dicres)
```
