```python
def bubblesort(arr):
    n = len(arr)
    for i in range(0, n):
        for j in range(0, n - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
    print(arr)
def Selectionsort(arr):
    n = len(arr)
    for i in range(0, n):
        minval = arr[i]
        minidx = i
        for j in range(i + 1, n):
            if arr[j] < minval:
                minval = arr[j]
                minidx = j
        arr[i], arr[minidx] = arr[minidx],arr[i]
    print(arr)
a = list(map(int, input().split()))
bubblesort(a)
Selectionsort(a)
```
